#Third part:
Feature Selection part

```python
#changing_Lasso_alpha (chi lasso score Analysis for feature selection)
def changing_Lasso_alpha(X,Y):
    st = time.time()
    X_train = X  
    y_train = Y
    
    X_train_scaled = scale(X_train)
    lasso = Lasso().fit(X_train_scaled, y_train)
    et = time.time()
    print('The score of the feature:')
    print(lasso.coef_)
    column_headers = list(df.columns)
    print('\nlist of best features in changing_Lasso_alpha methods:\n')
    for i in range(17):
        if lasso.coef_[i]>=0:#filter positive thresholding
            print(i,column_headers[i])
    elapsed_time = et - st
    print()
    print('changing_Lasso_alpha running time:', elapsed_time, 'seconds',)
```

# Nuova sezione

```python
#here the list of the best feature are extracted by changing lasso:
changing_Lasso_alpha(X,Y)
```

#Forth part
Prediction by Supervised Machine Learning

```python
#Shuffling and seperating data to train and test

df = shuffle(df)

#selected features
X = df[['Apps','Accept','Enroll','Top25perc','F.Undergrad','P.Undergrad',
'Top10perc','Outstate','Room.Board','PhD','S.F.Ratio']]

X = preprocessing.normalize(X)#features
Y  = df['Private'] #Target 

#-------------------------------------------KNN
def knn(X,y):
  #Create KNN Classifier
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1) 
  knn = KNeighborsClassifier(n_neighbors=5)

  #Train the model using the training sets
  knn.fit(X_train, y_train)

  #Predict the response for test dataset
  y_pred = knn.predict(X_test)

  print(" Knn model accuracy(in %):\n",100*metrics.accuracy_score(y_test, y_pred))

#------------------------------------------Decition Tree 
def decition_tree(X,y):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1) 
  # Create Decision Tree classifer object
  clf = DecisionTreeClassifier()

  # Train Decision Tree Classifer
  clf = clf.fit(X_train,y_train)

  #Predict the response for test dataset
  y_pred = clf.predict(X_test)
  print(" Decition tree model accuracy(in %):\n",100*metrics.accuracy_score(y_test, y_pred))

#------------------------------------------Logestic Regression:
def LR(x,y):
  X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
  sc_x = StandardScaler()
  xtrain = sc_x.fit_transform(X_train) 
  xtest = sc_x.transform(X_test)
  classifier = LogisticRegression(random_state = 0)
  classifier.fit(xtrain, y_train)
    
  y_pred = classifier.predict(xtest)
  print ("Logestic Regression model accuracy(in %):\n", 100*accuracy_score(y_test, y_pred))

#------------------------------------------Naive Biase Gaussian:
def NBG(X,y):

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)

    gnb = GaussianNB()
    gnb.fit(X_train, y_train)

    # making predictions on the testing set
    y_pred = gnb.predict(X_test)

    # comparing actual response values (y_test) with predicted response values (y_pred)

    print("Gaussian Naive Bayes model accuracy(in %):\n", metrics.accuracy_score(y_test, y_pred)*100)

#------------------------------------------Random Forest model
def RF(X,y):
   
    # define the model
    model = RandomForestClassifier()
    # evaluate the model
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    n_scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
    # report performance
    print("Random Forest model accuracy(in %):")
    print(100*mean(n_scores))

    return mean(n_scores)

#-------------------------------------------Support vector machine model
def SVM(X,y):

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
    #Create a svm Classifier
  clf = svm.SVC(kernel='linear') # Linear Kernel

  #Train the model using the training sets
  clf.fit(X_train, y_train)

  #Predict the response for test dataset
  y_pred = clf.predict(X_test)
  print("Support vector machine model accuracy(in %):\n",100*metrics.accuracy_score(y_test, y_pred))
```

```python
#knn:
knn(X,Y)
print('--------------------')
#Decition Tree:
decition_tree(X,Y)
print('--------------------')
#Logestic Regression:
LR(X,Y)
print('--------------------')
#Naiv Biase
NBG(X,Y)
print('--------------------')
#Random Forest
RF(X,Y)
print('--------------------')
#SVM
SVM(X,Y)
print('--------------------')
```
